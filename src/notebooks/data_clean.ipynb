## Importa√ß√µes

```python
import os
import shutil
from pathlib import Path

import kagglehub

import pandas as pd
```

## Defini√ß√£o do Dataframe

```python
print("=" * 60)
print("üìä CARREGANDO DATASET SPOTIFY TRACKS")
print("=" * 60)

os.makedirs("../datasets", exist_ok=True)

repo = "fcpercival/160k-spotify-songs-sorted"
filename = "data"
file_type = "csv"
dataset_path = f"../datasets/{filename}.{file_type}"

if not Path(dataset_path).exists():
    print("\n‚ö†Ô∏è Arquivo n√£o encontrado! Baixando via Kaggle Hub...")

    datasets_dir = Path("../datasets")
    datasets_dir.mkdir(exist_ok=True)

    try:
        path = kagglehub.dataset_download(repo)
        print(f"‚úÖ Dataset baixado para: {path}")

        source_file = None
        for file in Path(path).glob(f"*.{file_type}"):
            if filename in file.name:
                source_file = file
                print(f"‚úÖ Arquivo encontrado: {file.name}")
                break

        if source_file:
            destination = datasets_dir / f"{filename}.{file_type}"
            shutil.copy(source_file, destination)
            print(f"‚úÖ Arquivo copiado para: {destination}")
        else:
            print("‚ùå Arquivo espec√≠fico n√£o encontrado!")

    except Exception as e:
        print(f"‚ùå Erro ao fazer download: {e}")

print("‚úÖ Arquivo j√° est√° dispon√≠vel localmente.")
```

```python
print("\n" + "=" * 60)
print("üì• CARREGANDO DATASET")
print("=" * 60)

try:
    df = pd.read_csv(dataset_path)
    print(f"\n‚úÖ Carregado com sucesso!")
    print(f"   Total de registros: {len(df):,}")
    print(f"   Total de colunas: {len(df.columns)}")

except Exception as e:
    print(f"‚ùå Falha ao carregar: {e}")
    print("\n   Tentando carregar manualmente...")
```

# Limpeza dos dados

## Remo√ß√£o de features in√∫teis

```python
df_cleaned = df.copy()
df_cleaned.columns
```

```python
df_cleaned.drop(
    columns=["release_date", "key", "loudness", "mode", "liveness", "tempo"],
    inplace=True,
)
```

## Renomea√ß√£o das features

```python
print("\n" + "=" * 60)
print("üîß RENOMEANDO COLUNAS PARA LOWERCASE + SNAKE-CASE")
print("=" * 60)

df_renamed = df_cleaned.copy()

print(f"\nüìã Colunas originais ({len(df_renamed.columns)}):")
for i, col in enumerate(df_renamed.columns, 1):
    print(f"   {i:2d}. {col}")


def to_snake_case(name):
    name = name.lower()
    name = name.replace(" ", "_")
    name = name.replace("/", "_")
    name = "".join(c if c.isalnum() or c == "_" else "" for c in name)
    while "__" in name:
        name = name.replace("__", "_")
    return name


rename_dict = {col: to_snake_case(col) for col in df_renamed.columns}

df_renamed.rename(columns=rename_dict, inplace=True)

print("\n‚úÖ Renomea√ß√£o aplicada com sucesso!")
print(f"\nüìã Colunas renomeadas ({len(df_renamed.columns)}):")
for i, col in enumerate(df_renamed.columns, 1):
    print(f"   {i:2d}. {col}")

print("\nüîÑ Mapeamento de renomea√ß√£o:")
for old, new in rename_dict.items():
    if old != new:
        print(f"   {old:<30} ‚Üí {new}")
```

## An√°lise de nulos

```python
df_nulls = df_renamed.copy()
```

```python
print("\n" + "=" * 60)
print("üîç AN√ÅLISE DE VALORES NULOS (Incluindo -1)")
print("=" * 60)

df_nulls = df_renamed.copy()

nulos_nan = df_nulls.isnull().sum()
nulos_minus_one = (df_nulls == -1).sum()
nulos_total = nulos_nan + nulos_minus_one

total = df_nulls.shape[0]
porcentagem = (nulos_total / total) * 100

resultado = pd.DataFrame(
    {
        "NaN/None": nulos_nan,
        "Valores -1": nulos_minus_one,
        "Total Nulos": nulos_total,
        "Porcentagem (%)": porcentagem,
    }
)

print(f"\nüìä Total de registros: {total:,}\n")

colunas_com_nulos = resultado[resultado["Total Nulos"] > 0]

if len(colunas_com_nulos) == 0:
    print("‚úÖ Nenhum valor nulo encontrado!")
else:
    print(f"‚ö†Ô∏è Colunas com valores nulos ({len(colunas_com_nulos)}):\n")
    for idx, (col, row) in enumerate(colunas_com_nulos.iterrows(), 1):
        print(f"   {idx}. {col}")
        print(f"      NaN/None: {int(row['NaN/None']):,}")
        print(f"      Valores -1: {int(row['Valores -1']):,}")
        print(f"      Total: {int(row['Total Nulos']):,}")
        print(f"      Percentual: {row['Porcentagem (%)']:.2f}%")
        print()

print("=" * 60)
print("\nüìà Resumo completo:\n")
print(resultado.to_string())
```

```python
print("\n" + "=" * 60)
print("üóëÔ∏è  REMOVENDO VALORES -1 (NULOS)")
print("=" * 60)

# Contar registros antes
registros_antes = len(df_nulls)
print(f"\nüìä Registros antes da limpeza: {registros_antes:,}")

# Remover linhas onde qualquer coluna tem valor -1
df_nulls = df_nulls[(df_nulls != -1).all(axis=1)]

# Contar registros depois
registros_depois = len(df_nulls)
registros_removidos = registros_antes - registros_depois

print(f"üìä Registros ap√≥s limpeza: {registros_depois:,}")
print(f"üóëÔ∏è  Registros removidos: {registros_removidos:,}")
print(f"üìâ Percentual removido: {(registros_removidos / registros_antes) * 100:.2f}%")

# Verificar se ainda existem -1
if (df_nulls == -1).any().any():
    print("\n‚ö†Ô∏è  AVISO: Ainda existem valores -1 no dataset!")
    remaining = (df_nulls == -1).sum()
    print(f"   {remaining[remaining > 0]}")
else:
    print("\n‚úÖ Todos os valores -1 foram removidos com sucesso!")

print("\n" + "=" * 60)
```
